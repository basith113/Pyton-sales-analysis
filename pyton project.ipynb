```python
#import library

import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
#load  data

df=pd.read_csv(r"C:\Users\abdul\OneDrive\Desktop\文档\Downloads\Diwali Sales Data.csv",encoding='unicode_escape')
```

```python
#verify the Size
df.shape
```

```python
#Read the top 10 rows
df.head(10)
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({0: 'Married', 1: 'Unmarried'})
```

```python
df.info()
```

```python
#drop unrelated or Blank Column

df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
#count the null values in each column
pd.isnull(df).sum()
```

```python
#drop the null values
df.dropna(inplace=True)
```

```python
#Change the data type
df['Amount']=df['Amount'].astype('int')
```

```python

# count the total rows group by gender

ax = sns.countplot(x='Gender', data=df)

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Gender wise amount

df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Create the bar plot
ax = sns.barplot(x='Gender', y='Amount', data=sales_gen)

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
# Create the countplot and assign it to 'ax'
ax = sns.countplot(data=df, x='Age Group')

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#gender wise age group

ax=sns.countplot(data=df,x='Age Group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Grouping the data by 'Age Group' and summing the 'Amount'
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Create a barplot and assign it to 'ax'
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age)

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Grouping the data by 'State' and summing the 'Orders'
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Getting top 10 states
top_10_states = sales_state.head(10)

# Setting the figure size
sns.set(rc={'figure.figsize':(15, 5)})

# Creating the bar plot for top 10 states
ax = sns.barplot(data=top_10_states, x='State', y='Orders')

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Grouping the data by 'State' and summing the 'Orders'
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Getting top 10 states
top_10_states = sales_state.head(10)

# Setting the figure size
sns.set(rc={'figure.figsize':(15, 5)})

# Creating the bar plot for top 10 states
ax = sns.barplot(data=top_10_states, x='State', y='Amount')

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Count plot for 'Marital_Status'
ax = sns.countplot(data=df, x='Marital_Status')

# Setting the figure size
sns.set(rc={'figure.figsize':(5, 5)})

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Grouping the data by 'Marital_Status' and 'Gender' and summing the 'Amount'
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount')

# Setting the figure size
sns.set(rc={'figure.figsize':(10, 5)})

# Creating the bar plot with 'Marital_Status' on the x-axis, 'Amount' on the y-axis, and 'Gender' as hue
ax=sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender')

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Setting the figure size
sns.set(rc={'figure.figsize':(20, 5)})

# Sorting the data by 'Occupation' counts
occupation_counts = df['Occupation'].value_counts().reset_index()
occupation_counts.columns = ['Occupation', 'Count']

# Creating the count plot for 'Occupation' after sorting
ax = sns.countplot(data=df, x='Occupation', order=occupation_counts['Occupation'])

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Grouping the data by 'State' and summing the 'Orders'
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Getting top 10 states
top_10_states = sales_state.head(10)

# Setting the figure size
sns.set(rc={'figure.figsize':(15, 5)})

# Creating the bar plot for top 10 states
ax = sns.barplot(data=top_10_states, x='Occupation', y='Amount')

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Setting the figure size
sns.set(rc={'figure.figsize':(20, 5)})

# Sorting the data by 'Occupation' counts
occupation_counts = df['Product_Category'].value_counts().reset_index().head(10)
occupation_counts.columns = ['Product_Category', 'Count']

# Creating the count plot for 'Occupation' after sorting
ax = sns.countplot(data=df, x='Product_Category', order=occupation_counts['Product_Category'])

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Grouping the data by 'State' and summing the 'Orders'
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

# Getting top 10 states
top_10_states = sales_state.head(10)

# Setting the figure size
sns.set(rc={'figure.figsize':(20, 5)})

# Creating the bar plot for top 10 states
ax = sns.barplot(data=top_10_states, x='Product_Category', y='Amount')

# Adding labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)
```
